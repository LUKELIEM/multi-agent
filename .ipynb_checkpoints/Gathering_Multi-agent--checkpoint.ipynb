In this notebook, we progressively move towards multi-agent reinforcement learning:

* 1st iteration - Policy Gradient agent vs 1 Random Agent
* 2nd Iteration - Policy Gradient agent vs 1+ Random Agent
* 3rd iteration - Iteratively train Policy Gradient agent against a trained PG agent

## Policy Gradient Agent

The agent's policy is a simple 2-layer NN. The code below only allows one agent to be trained in a Gathering environment with the other agents being random agents.

Note that the weights are saved and loaded from a fixed single file.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable


class Policy(nn.Module):
    
    # an index parameter may be needed to allow for multiple learning agents
    def __init__(self, observation_size):
        super(Policy, self).__init__()
        self.affine1 = nn.Linear(observation_size, 128)
        self.affine2 = nn.Linear(128, 8)

        self.saved_actions = []
        self.rewards = []

    def forward(self, x):
        x = F.relu(self.affine1(x))
        action_scores = self.affine2(x)
        return F.softmax(action_scores)

    # The weights should be allowed to be saved into and loaded different model files
    def save_weights(self):
        torch.save(self.state_dict(), 'model.pkl')   

    def load_weights(self):
        self.load_state_dict(torch.load('model.pkl'))

    def select_action(self, state):
        state = torch.from_numpy(state).float().unsqueeze(0)
        probs = self(Variable(state))
        action = probs.multinomial()
        self.saved_actions.append(action)
        return action.data[0, 0]
```

## Iteration 2 - PG Agent vs 2 Random Agents

We get the code to accept more than 2 agents. And it works after making some change to GatheringEnv() Class.

```python
env = GatheringEnv(n_agents=3, map_name='default')  # 

# Env API is similar to that of OpenAI Gym
state_n = env.reset()
env.render()

# Load previously trained policy agent
policy = Policy(env.state_size)
policy.load_weights()

n_tags = 0 # count number of times learning agent tag the random agent
n_steps = 300

# Render for 1000 steps
for _ in range(n_steps):
    action = policy.select_action(state_n[0])
    state_n, reward_n, done_n, info_n = env.step([
        action,                           # Agent 1 is policy AI
        random.randrange(0, 8),           # Agent 2 is random
        random.randrange(0, 8)            # Agent 3 is random
    ])
    
    if action is 6:
        n_tags += 1
        
    if any(done_n):
        break
    env.render()
    time.sleep(1/30)  # Change speed of video rendering

env.close()  # Close the rendering window
print ("Aggressiveness is {:.2f}".format(n_tags/n_steps))
```

## Iteration 3 - PG Agent vs Previously-trained PG Agent

To get the code to accept multiple trained agents, the Policy class must be changed to accept agent index.

The code now expects each agent to have its own set of trained weights:

* agent-1-model.pkl
* agent-2-model.pkl
* etc....

```python
import os

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable


class Policy(nn.Module):
    
    # an index parameter may be needed to allow for multiple learning agents
    def __init__(self, observation_size, agent_idx=1):
        super(Policy, self).__init__()
        self.affine1 = nn.Linear(observation_size, 128)
        self.affine2 = nn.Linear(128, 8)

        self.saved_actions = []
        self.rewards = []
        self.idx = agent_idx   # This allows multiple learning agents


    def forward(self, x):
        x = F.relu(self.affine1(x))
        action_scores = self.affine2(x)
        return F.softmax(action_scores)

    # The weights should be allowed to be saved into and agent-indexed model files
    def save_weights(self):
        file_name = 'agent-'+str(self.idx)+'-model.pkl'
        torch.save(self.state_dict(), file_name)   

    def load_weights(self):
        file_name = 'agent-'+str(self.idx)+'-model.pkl'
        if not os.path.exists(file_name):
                raise ValueError('map not found: ' + file_name)
        self.load_state_dict(torch.load(file_name))

    def select_action(self, state):
        state = torch.from_numpy(state).float().unsqueeze(0)
        probs = self(Variable(state))
        action = probs.multinomial()
        self.saved_actions.append(action)
        return action.data[0, 0]
```

## Multi-agents with previously trained weights

When we load 3 PG agents with identical weights which was the result of training a PG agent against a random agent:

* Both agent 1 and 2 are loaded with SA_ep3000_lr1e-3_default_r440.pkl

They appear to act as expected - being nearly mirror-image of each other.

```python
# There will be 3 agents - all AI agents, no random agents
num_agents = 3
num_ai_agents = 3

# Data structure for AI agents (agents will form their own Class later on)
ai_agents = []
actions = []
tags = []
rewards = []

env = GatheringEnv(n_agents=num_agents, map_name='default')

# Env API is similar to that of OpenAI Gym
state_n = env.reset()
env.render()

# Load AI agent with trained weights
for i in range(num_ai_agents):
    ai_agents.append(Policy(env.state_size, i+1))
    ai_agents[i].load_weights()

# Initialize AI agent data
for i in range(num_ai_agents):
    actions = [0 for i in range(num_ai_agents)]
    tags = [0 for i in range(num_ai_agents)]
    rewards = [0 for i in range(num_ai_agents)]

n_steps = 300

# Render for 1000 steps
for step in range(n_steps):
    # Load AI agent with trained weights
    for i in range(num_ai_agents):
        actions[i] = ai_agents[i].select_action(state_n[i])
        if actions[i] is 6:
            tags[i] += 1   # record a tag for accessing aggressiveness

    if step % 10 == 0:
        print (actions)    
            
    state_n, reward_n, done_n, info_n = env.step(actions)

    for i in range(num_ai_agents):
        rewards[i] += reward_n[i]    # Accumulate rewards for each agent
        
    if any(done_n):
        break
    env.render()
    time.sleep(1/30)  # Change speed of video rendering

env.close()  # Close the rendering window

for i in range(num_ai_agents):
    print ("Agent{} aggressiveness is {:.2f}".format(i+1, tags[i]/n_steps))
    print ("Agent{} reward is {:d}".format(i, rewards[i]))
```

## REINFORCE

We will now attempt to train a PG agent in a multi-agent environment with a previously trained PG agent.

```python
# Adapted from https://github.com/pytorch/examples/blob/2dca104/reinforcement_learning/reinforce.py
# Licensed under BSD 3-clause: https://github.com/pytorch/examples/blob/2dca10404443ce3178343c07ba6e22af13efb006/LICENSE

import random
from itertools import count

from env import GatheringEnv

import numpy as np

import torch
import torch.autograd as autograd
import torch.optim as optim

# We remove the argparse code and replace with hard parameters
gamma = 0.99
seed = 543
render=False
log_interval=20
target_reward = 500
                 
# Start the Gathering environment
env = GatheringEnv(n_agents=2, map_name='default')
env.seed(seed)
torch.manual_seed(seed)

# Load the learning agent and train from scratch
policy_agent1 = Policy(env.state_size, 1)
# policy_agent1.load_weights()  

# Load previously trained policy agent
policy_agent2 = Policy(env.state_size, 2)
policy_agent2.load_weights()

optimizer = optim.Adam(policy_agent1.parameters(), lr=1e-3)   # Only agent 1 will learn

# This is the task list to finish a game episode
def finish_episode():
    """ 
    Based on REINFORCE, policy gradient is calculated at the end of an episode.
    It is then used to update the Policy's weights
    """
    
    # Out code only accommodates 1 learning agent.
    # Reward for each time step is stored in the list policy.rewards[] --> r(t)
    R = 0  # noqa
    rewards = []
    for r in policy_agent1.rewards[::-1]:
        R = r + gamma * R  # noqa
        rewards.insert(0, R)
    rewards = torch.Tensor(rewards)
    rewards = (rewards - rewards.mean()) / (rewards.std() + np.finfo(np.float32).eps)
    
    # Calculate policy gradient Σ_t[r(t) ∇_θ log(π_θ(s_t,a_t))
    for action, r in zip(policy_agent1.saved_actions, rewards):
        action.reinforce(r)     # Note that action.reinforce has been deprecated since torch V0.3
    optimizer.zero_grad()
    autograd.backward(policy_agent1.saved_actions, [None for _ in policy_agent1.saved_actions])
    optimizer.step()
    
    del policy_agent1.rewards[:]
    del policy_agent1.saved_actions[:]


running_reward = None
best_reward = None  # Keep track of best running average for storing best_model
for i_episode in range(1,1001):   # For each game episode
    
    # Some basic initialization
    state = env.reset()
    episode_reward = 0
    
    for t in range(1000):  # Don't infinite loop while learning
        
        # 1. learning agent and trained agent select their action
        action_agent1 = policy_agent1.select_action(state[0])
        action_agent2 = policy_agent2.select_action(state[1])
        
        # 2. environment accepts action from learning agent and action from 2nd agent
        # and doles out next state, reward and if the episode is done
        state_n, reward_n, done_n, _ = env.step([action_agent1, action_agent2])
        # Only save agent 1's state and reward for learning
        state = state_n
        done = done_n[0]
        
        if render:   # render learning if set to True
            env.render()
            
        # Store rewards of agent 1 into array for REINFORCE    
        policy_agent1.rewards.append(reward_n[0])
        episode_reward += reward_n[0]
        
        if done:
            break

    # Update reward statistics
    if running_reward is None:
        running_reward = episode_reward
    running_reward = running_reward * 0.99 + episode_reward * 0.01
    
    finish_episode()    # Finish up game episode
        
    if i_episode % log_interval == 0:
        print('Episode {}\tLast reward: {:5d}\tAverage reward: {:.2f}'.format(
            i_episode, episode_reward, running_reward))
        # Only save NN weights if running_reward is better than best_reward
        if best_reward is None:
            best_reward = running_reward
            policy_agent1.save_weights()   # Save NN weights for agent 1 if this is the first save
        else:
            if running_reward > best_reward:
                policy_agent1.save_weights()   # Save NN weights for agent 1 only if it performs better
        
    if running_reward > target_reward:    # Rather weird to 
        print("Solved! Running reward is now {} and "
              "the last episode received {} reward!".format(running_reward, episode_reward))
        policy_agent1.save_weights()   # Save NN weights for agent 1
        env.close()  # Close the rendering window
        break
```

## Play - Newly Trained PG1 vs Trained PG2

The following 2 agents are pitted together:

* Agent 1 (PG agent trained against static PG agent for 1000 episodes)
    * MA_ep1000_lr1e13_default.pkl
    
* Agent 2 (PG agent trained against random agent for 3000 episodes)
    * SA_ep3000_lr1e-3_default_r440.pkl 
    
As expected, Agent 1 completely dominates, outtagging agent 2 out by being more aggressive (13% vs 10% tag rate) and eating 196 food units.

```python
env = GatheringEnv(n_agents=2, map_name='default')  # 

# Env API is similar to that of OpenAI Gym
state_n = env.reset()
env.render()

# Load previously trained policy agent
policy_agent1 = Policy(env.state_size, 1)
policy_agent1.load_weights()

# Load previously trained policy agent
policy_agent2 = Policy(env.state_size, 2)
policy_agent2.load_weights()

agent1_reward = 0
agent2_reward = 0
agent1_tags = 0 # count number of times learning agent tag the other agent
agent2_tags = 0 # count number of times learning agent tag the other agent

n_steps = 500

# Render for 1000 steps
for _ in range(n_steps):
    action_agent1 = policy_agent1.select_action(state_n[0])
    action_agent2 = policy_agent2.select_action(state_n[1])
    state_n, reward_n, done_n, info_n = env.step([
        action_agent1,                    # Agent 1 is policy AI
        action_agent2,                    # Agent 2 is policy AI
        # random.randrange(0, 8)            # Agent 3 is random
    ])
    
    if action_agent1 is 6:
        agent1_tags += 1
        
    if action_agent2 is 6:
        agent2_tags += 1
        
    agent1_reward += reward_n[0]
    agent2_reward += reward_n[1]
        
    if any(done_n):
        break
    env.render()
    time.sleep(1/30)  # Change speed of video rendering

env.close()  # Close the rendering window
print ("Agent1 aggressiveness is {:.2f}".format(agent1_tags/n_steps))
print ("Agent2 aggressiveness is {:.2f}".format(agent2_tags/n_steps))
print ("Agent1 reward is {:d}".format(agent1_reward))
print ("Agent2 reward is {:d}".format(agent2_reward))
```
